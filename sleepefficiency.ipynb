```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

df = pd.read_csv("Sleep_Efficiency.csv")

corr_matrix = df.info()
print(corr_matrix)

#cleaning
sleep = df.dropna(axis=0, how='any')
```

```python
droplist = ['ID', 'Bedtime', 'Wakeup time', 'REM sleep percentage', 'Deep sleep percentage', 'Light sleep percentage', 'Awakenings']

sleep.drop(droplist, axis=1, inplace=True)

sleep.info()

plt.rcParams['font.size'] = 5
pd.plotting.scatter_matrix(sleep, alpha=1)
```

```python
#sleep.drop(["Age", "Sleep duration", "Sleep efficiency"], axis=1, inplace = True)

sleep.plot.scatter(x="Exercise frequency", y = "Sleep efficiency")
```

```python
r = np.corrcoef(sleep["Exercise frequency"], sleep["Sleep duration"])
print(r)
```
